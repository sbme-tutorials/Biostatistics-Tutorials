```python
import pandas as pd
healthy = pd.read_csv('data/lusc-rsem-fpkm-tcga_paired.txt', sep='\t')
cancer = pd.read_csv('data/lusc-rsem-fpkm-tcga-t_paired.txt', sep='\t')
```

## Picking individual Gene from healthy and cancer cases

```python
# Picking first gene fron healthy set
G1_h = healthy.iloc[0, 2:]
G1_h
```

```python
# Picking first gene fron cancer set
G1_c = cancer.iloc[0, 2:]
G1_c
```

## Correlation

### 1. Pearson correlation 

```python
from scipy.stats import pearsonr
r, _ = pearsonr(G1_h, G1_c)
print("Pearson Correlation (r) =  ", r)
```

### 2. Spearman correlation  

```python
from scipy.stats import spearmanr
r, _ = spearmanr(G1_h, G1_c)
print("Spearman Correlation (r) =  ", r)
```

## Hypothesis test

### 1. Samples are paired (related)

```python
from scipy.stats import ttest_rel
p_val = ttest_rel(G1_h, G1_c).pvalue
print("p-value = ", p_val)
```

### 2. Samples are independant 

```python
from scipy.stats import ttest_ind
p_val = ttest_ind(G1_h, G1_c).pvalue
print("p-value = ", p_val)
```

## Multiple tests correction (FDR Correction)

```python
from statsmodels.stats.multitest import multipletests
import numpy as np 
```

```python
# After getting all p values for all genes you will have a list of p values
# so we can apply fdr correction 
# Lets assume we have 10 p values from 0.005 to 0.1
gene_names = ['G' + str(i) for i in range(1, 11)]
p_values = list(np.random.uniform(0.001, 0.07, 10))
corrected_p_values = multipletests(p_values, alpha=0.05, method='fdr_bh')[1]
```

```python
significance_genes = pd.DataFrame({'Gene_name':gene_names, 'p-values':p_values, 'p-values_fdr':corrected_p_values})
significance_genes
```

```python
significance_genes['significance:p_vlaue'] = significance_genes['p-values'].apply(lambda x: x < 0.05)
significance_genes['significance:p_vlaue_fdr'] = significance_genes['p-values_fdr'].apply(lambda x: x < 0.05)
significance_genes
```

```python
# Get significant genes after fdr correction
diffrentially_genes = significance_genes[significance_genes['significance:p_vlaue_fdr']== True]
diffrentially_genes
```

```python
# To get gene names 
diffrentially_genes['Gene_name'].to_list()
```
